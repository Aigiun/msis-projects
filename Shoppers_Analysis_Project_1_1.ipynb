```python
shopping_df_mod["Victoria's Secret"] = shopping_df_mod.apply(VictoriaSecret, axis = 'columns')

shopping_df_mod["Llbean"] = shopping_df_mod.apply(Llbean, axis = 'columns')

shopping_df_mod["Landsend"] = shopping_df_mod.apply(LandSend, axis = 'columns')

shopping_df_mod["OldNavy"] = shopping_df_mod.apply(OldNavy, axis = 'columns')

shopping_df_mod["JCPenney"] = shopping_df_mod.apply(JCPenney, axis = 'columns')

shopping_df_mod["GAP"] = shopping_df_mod.apply(GAP, axis = 'columns')

shopping_df_mod["Blair"] = shopping_df_mod.apply(Blair, axis = 'columns')

shopping_df_mod["AE"] = shopping_df_mod.apply(AE, axis = 'columns')

shopping_df_mod["Kohls"] = shopping_df_mod.apply(Kohls, axis = 'columns')

shopping_df_mod["WomanWithin"] = shopping_df_mod.apply(WomanWithin, axis = 'columns')

shopping_df_mod
```

```python
shopping_df_final = shopping_df_mod.drop(columns = ['site1', 'site2', 'site3', 'site4', 'site5'])
shopping_df_final
```

```python
# Create frequent itemsets
itemsets = apriori(shopping_df_final, min_support = 0.12, use_colnames = True)
itemsets
```

```python
rules = association_rules(itemsets, metric='confidence', min_threshold=0.30)
rules
```
